```python
names = []
for name in tmp2.columns:
  if name == '행정구역':
    # print(name)
    names.append(name)
    continue
  # print( pd.to_datetime(name.split('_')[0], format='%Y년%m월') )
  names.append(pd.to_datetime(name.split('_')[0], format='%Y년%m월'))
```
