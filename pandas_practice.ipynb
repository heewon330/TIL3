```python
# 지점별 몸무게 평균 -피벗 사용
# aggfunc의 기본값이 평균이라 생략
pd.pivot_table(rawData,index='지점',values='몸무게')
```

```python
# apply - 판다스에서 사용 가능한 반복문
# 모든 행과 열에 대해 동일한 함수를 반복적으로 적용 가능

def func(x): 
  print('함수가 호출되었습니다')
  print('x:{}'.format(x))

#apply로 호출된 함수는 꼭 return 해줘야 함
  return

# 일반적으로 데이터 프레임의 각 컬럼들이 시리즈 형태로 전달됨
rawData.apply(func)
```

```python
# 행을 전달하고자 한다면?
def func(x): 
  print('함수가 호출되었습니다')
  print('x:{}'.format(x))

#axis=0 : 열, axis=1 : 행
rawData.apply(func, axis=1)
```

```python
# 지점별로 지역을 새로 만들기

def func(x):
  if x['지점'] =='관악구' : return '강서'
  elif x['지점'] =='여의도' : return '강서'
  elif x['지점'] =='강남구' : return '강동'
  elif x['지점'] =='서초구' : return '강동'

#apply가 리턴하는 값들도 하나의 시리즈가 됨

rawData.apply(func,axis=1)
```

```python
#새로운 컬럼 추가
rawData['지역']=rawData.apply(func,axis=1)
rawData
```

```python
# 프레임 합치기 join
user_device=pd.read_csv('/content/drive/MyDrive/멀티캠퍼스/data/user_device.csv')
user_usage=pd.read_csv('/content/drive/MyDrive/멀티캠퍼스/data/user_usage.csv')
```

```python
user_usage.head(2) #휴대폰 사용 데이터
```

```python
user_device.head(2) #휴대폰기기
```

```python
display(len(user_usage))
display(len(user_device)) #조인된 자료의 수는 확 줄 것
```

```python
# inner join
# merge 함수 통해서 가능

pd.merge(left=user_usage, right=user_device, on='use_id')
```

```python
# left join 
#결측치 발생 가능 -right에 없는 자료는 결측치로~
#240개 데이터(레프트 데이터 수와 동일)

pd.merge(left=user_usage, right=user_device, on='use_id',how='left')
```

```python
# full outer join
# 양쪽 모두

pd.merge(left=user_usage, right=user_device, on='use_id',how='outer')
```
