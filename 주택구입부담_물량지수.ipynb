# 주택구입부담지수란?
- 주택구입부담지수 = 대출상환가능소득 / 중간가구소득(월) *100 = (원리금 상환액 / DTI) / 중간가구소득(월) * 100
 - 원리금상환액은 LTV 47.9%, DTI 25.7%, 만기 20년 원리금균등 상환대출의 매월 상환액
 - LTV : 주택담보가치 대비 대출이 가능한 금액. 쉽게 말하자면 시세 대비 몇 프로까지 주택담보대출을 받을 수 있는지 확인하는 것
 - DTI : 대출이용자의 연소득대비 대출 상환액. DTI 비율은 연간 소득을 연간 대출상환액으로 나눈 금액에 100을 곱해서 계산
- 중간소득가구가 표준대출을 받아 중간가격 주택을 구입하는 경우의 상환부담을 나타내는 지수로서, 지수가 낮을수록 주택구입부담이 완화됨을 의미
- 주택담보대출 상환으로 가구소득의 약25%를 부담하면 주택구입부담지수는 100으로 산출됨

# 주택구입물량지수
- 주택구입물량지수 = 중위소득 가구가 구입가능한 주택물량 / 전체 주택물량 *100
- 중위가구가 구입 가능한 주택수의 비율(0~100)로 주택구입능력을 측정하는 지수로서, 수치가 높을수록 중위소득 가구 구입 가능한 주택물량이 많다는 의미로, 100은 중위소득 가구가 주택을 100% 구입 가능함을 의미
